## Multi variate analysis

```python
insurance_data['Processed_Status'] = insurance_data['CLAIM_STATUS'].apply(lambda x: 'A' if x == 'A' else 'D')

# aggregate these counts by state.
status_counts_by_state = insurance_data.groupby(['STATE', 'Processed_Status']).size().unstack(fill_value=0)
```

```python
# Plotting the data
status_counts_by_state.plot(kind='bar', stacked=True, figsize=(14, 7), color=['green', 'red'])

plt.title('Accept and Reject Counts by State')
plt.xlabel('State')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Claim Outcome')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Select only numerical columns for calculating quartiles
numeric_columns = insurance_data.select_dtypes(include=[np.number])

# Calculate the first quartile (Q1) and third quartile (Q3) for numerical columns
Q1 = numeric_columns.quantile(0.25)
Q3 = numeric_columns.quantile(0.75)

# Calculate the interquartile range (IQR) for numerical columns
IQR = Q3 - Q1

# Identify outliers based on the IQR method for numerical columns
outliers = (numeric_columns < (Q1 - 1.5 * IQR)) | (numeric_columns > (Q3 + 1.5 * IQR))

# Count the number of outliers in each numerical column
num_outliers = outliers.sum()

print("Number of outliers in each numerical column:")
print(num_outliers)
```
