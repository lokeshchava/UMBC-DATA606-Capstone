```python
import pandas as pd

# Load datasets
employee_data = pd.read_csv("employee_data.csv")
vendor_data = pd.read_csv("vendor_data.csv")
insurance_data = pd.read_csv("insurance_data.csv")
```

```python
import os

file_paths = ['employee_data.csv','vendor_data.csv','insurance_data.csv']

# Iterate over each file path
for file_path in file_paths:
    # Check if the file exists
    if os.path.exists(file_path):
        # Get the size of the file in bytes
        file_size_bytes = os.path.getsize(file_path)

        # Convert bytes to kilobytes (KB), megabytes (MB)
        file_size_kb = file_size_bytes / 1024
        file_size_mb = file_size_kb / 1024

        # Print the file size in different units
        print(f"File: {file_path}")
        print("KB:", round(file_size_kb, 2))
        print("MB:", round(file_size_mb, 2))
        print()
    else:
        print(f"File '{file_path}' not found.")
```

```python
#dataset size
print("Employee Data Statistics:")
print(employee_data.shape)

print("\nVendor Data Statistics:")
print(vendor_data.shape)

print("\nInsurance Claims Data Statistics:")
print(insurance_data.shape)
```

```python
#timestamp period for employee dataset

if 'DATE_OF_JOINING' in employee_data.columns:
    # Convert 'DATE_OF_JOINING' column to datetime dtype
    employee_data['DATE_OF_JOINING'] = pd.to_datetime(employee_data['DATE_OF_JOINING'])

    # Find the minimum and maximum timestamps
    min_timestamp = employee_data['DATE_OF_JOINING'].min()
    max_timestamp = employee_data['DATE_OF_JOINING'].max()

    # Print the time period
    print("Time period for the employee dataset:")
    print("From:", min_timestamp.strftime('%Y-%m-%d %H:%M:%S'))
    print("To:", max_timestamp.strftime('%Y-%m-%d %H:%M:%S'))
else:
    print("No timestamp column found in the dataset.")
```

```python
#timestamp period for insurance dataset

if 'TXN_DATE_TIME' in insurance_data.columns:
    # Convert 'TXN_DATE_TIME' column to datetime dtype
    insurance_data['TXN_DATE_TIME'] = pd.to_datetime(insurance_data['TXN_DATE_TIME'])

    # Find the minimum and maximum timestamps
    min_timestamp = insurance_data['TXN_DATE_TIME'].min()
    max_timestamp = insurance_data['TXN_DATE_TIME'].max()

    # Print the time period
    print("Time period for the insurance dataset:")
    print("From:", min_timestamp.strftime('%Y-%m-%d %H:%M:%S'))
    print("To:", max_timestamp.strftime('%Y-%m-%d %H:%M:%S'))
else:
    print("No timestamp column found in the dataset.")
```

```python
# identify categorical and numerical variables

def identify_variables(data):
    numerical_vars = data.select_dtypes(include=['number']).columns.tolist()
    categorical_vars = data.select_dtypes(include=['object']).columns.tolist()
    return numerical_vars, categorical_vars

# Identify variables for each dataset
employee_numerical, employee_categorical = identify_variables(employee_data)
vendor_numerical, vendor_categorical = identify_variables(vendor_data)
insurance_numerical, insurance_categorical = identify_variables(insurance_data)

# Print results for Employee Dataset
print("Employee Data:")
print("Numerical Variables:", employee_numerical)
print("Categorical Variables:", employee_categorical)
print()
```

```python
# Print results for Vendor Dataset
print("Vendor Data:")
print("Numerical Variables:", vendor_numerical)
print("Categorical Variables:", vendor_categorical)
print()
```
